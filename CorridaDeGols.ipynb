```python
from google.colab import drive
import pandas as pd
import csv
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Analise de gols/results.csv', parse_dates=['date'])
flags = pd.read_csv('/content/drive/MyDrive/Analise de gols/countries-fifa-flags.csv')
df
```

```python
df.info()
```

```python
resultado_Copas = df[df['tournament'] == 'FIFA World Cup']
resultado_Copas
```

```python
resultado_Copas['year'] = resultado_Copas['date'].dt.year
```

```python
home_score = resultado_Copas[['home_team','home_score', 'year' ]].rename({'home_team' : 'team','home_score': 'score'}, axis = 'columns')
away_score = resultado_Copas[['away_team','away_score', 'year' ]].rename({'away_team' : 'team','away_score': 'score'}, axis = 'columns')
```

```python
scores_Copas = pd.concat([home_score, away_score]).reset_index(drop=True).fillna(0)
scores_Copas
```

```python
scores_fifa_year = scores_Copas.groupby(['team','year']).score.sum().unstack().fillna(0).cumsum(axis = 1).reset_index()
```

```python
scores_fifa_year.merge(flags, left_on = 'team', right_on ='country', how='left').to_csv('Gols-Fifa-WC.csv')
```
